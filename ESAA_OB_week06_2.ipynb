### 5.1.1 합성곱층의 필요성

이미지, 영상 처리에 유용  

- 배열을 펼쳐서
- 각 픽셀에 가중치 곱
- 은닉층으로 전달.  

단 데이터의 공간적 구조를 무시하게 됨 -> 합성곱층 필요


## 5.1.2 합성곱 신경망 구조

다차원 배열 데이터 처리에 특화  

계층 5개로 구성

1. 입력층

이미지: 높이, 너비, 채널 값 갖는 3차원 데이터  
채널: gray scale이면 1, RGB면 3 값 가짐.


2. 합성곱층

커널/필터 이용해 입력 데이터에서 특성 추출  
-> '특성 맵' 추출, 크기가 줄어듦

- gray scale 이미지

일반적으로 커널은 3x3, 5x5 크기.  
스트라이드(stride)라는 지정된 간격에 따라 순차적 이동.  
입력 이미지&필터 포개 대응되는 숫자끼리 곱한 후 모두 더함  

- RGB 이미지

필터 채널이 3(단, 필터 개수 3개 아니고 1개!!)  
RGB 각각에 서로 다른 가중치로 합성곱 적용 후 결과 더해줌.

필터가 2개 이상이면  
필터 각각이 특성 추출 결과의 채널이 됨.



3. 풀링층 - 2,3번 반복으로 주요 특성 벡터 추출

- 특성 맵의 차원을 다운 샘플링하여 연산량 감소
- 주요 특성 벡터 추출(효과적 학습)

연산 방식
- **최대 풀링**: 대상 영역에서 최댓값 추출
- 평균 풀링: 대상 영역에서 평균 반환 but 중요한 가중치 갖는 값 특성이 희미해질 수 있음.  

4. 완전연결층 - 1차원 벡터로 변환



5. 출력층 - 활성화함수 사용해 최종 결과 출력

소프트맥스 활성화 함수 사용  
입력받은 값을 0~1 사이의 값으로 출력 - 이미지가 각 레이블에 속할 확률 값.  
가장 높은 확률 값 갖는 레이블이 최종 값으로 선정  



### 5.1.3 1D, 2D, 3D 합성곱

이동하는 방향의 수 / 출력 형태에 따라 분류

#### 1D 합성곱

필터가 시간을 축으로 좌우로만 이동 가능

#### 2D 합성곱

필터가 방향 두 개로 움직임.  

#### 3D 합성곱

필터가 방향 세 개로 움직임.  
필터(k,k,d)이고 입출력은 (W,H,L) 형태.  
d < L 유지하는 것이 중요.  

#### 3D 입력을 갖는 2D 합성곱

필터 길이 = 입력 채널 길이일 때,  
두 방향으로 움직임

#### 1X1 합성곱

3D 형태 입력(W, H, L) + 필터(1,1,L) -> 출력(W,H)
