```python
import pandas as pd
import glob, os
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_colwidth', 700)

# ëë í°ë¦¬ ì£¼ìë¡ ë¤ì ì¤ì 
path = r'/content/drive/MyDrive/EWHA/ESAA/dataset/topics'
# pathë¡ ì§ì í ëë í°ë¦¬ ë°ì ìë ëª¨ë  .data íì¼ë¤ì íì¼ëªì ë¦¬ì¤í¸ë¡ ì·¨í©
all_files = glob.glob(os.path.join(path, "*.data"))
filename_list = []
opinion_text = []

# ê°ë³ íì¼ë¤ì íì¼ëªì filename_list ë¦¬ì¤í¸ë¡ ì·¨í©
# ê°ë³ íì¼ë¤ì íì¼ ë´ì©ì DataFrame ë¡ë© í ë¤ì stringì¼ë¡ ë³ííì¬ opinion_textë¦¬ì¤í¸ë¡ ì·¨í©
for file_ in all_files:
  # ê°ë³ íì¼ì ì½ì´ì DFë¡ ìì±
  df = pd.read_table(file_, index_col=None, header=0, encoding='latin1')
  # ì ë ê²½ë¡ë¡ ì£¼ì´ì§ íì¼ëªì ê°ê³µ. Linuxìì ìí ììë ìë \\ë¥¼ /ë¡ ë³ê²½.
  # ë§¨ ë§ì§ë§ .data íì¥ìë ì ê±°.
  filename_ = file_.split('/')[-1]
  filename = filename_.split('.')[0]

  # íì¼ëª ë¦¬ì¤í¸ì íì¼ë´ì© ë¦¬ì¤í¸ì íì¼ëªê³¼ íì¼ ë´ì©ì ì¶ê°.
  filename_list.append(filename)
  opinion_text.append(df.to_string())

# íì¼ëª ë¦¬ì¤í¸ì íì¼ ë´ì© ë¦¬ì¤í¸ë¥¼ DataFrameì¼ë¡ ìì±
document_df = pd.DataFrame({'filename': filename_list, 'opinion_text': opinion_text})
document_df.head()
```

> TF-IDF ííë¡ í¼ì² ë²¡í°í

TfidfVectorizer: Lemmatization ê°ì ì´ê·¼ ë³í ì§ì  ì§ìX  
-> tokenizer ì¸ìì ì»¤ì¤í ì´ê·¼ ë³í í¨ìë¥¼ ì ì©í´ ì´ê·¼ ë³í ìí ê°ë¥

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

def LemNormalize(text):
  lemmatizer = WordNetLemmatizer()
  return [lemmatizer.lemmatize(token) for token in word_tokenize(text)]
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(tokenizer=LemNormalize, stop_words='english', \
                             ngram_range=(1,2), min_df=0.05, max_df=0.85)

#opinion_text ì¹¼ë¼ê°ì¼ë¡ feature vectorization ìí
feature_vect = tfidf_vect.fit_transform(document_df['opinion_text'])
```

> êµ°ì§í ìí

K-íê·  ì ì©

```python
from sklearn.cluster import KMeans

# 5ê° ì§í©ì¼ë¡ êµ°ì§í ìí. ìì ë¥¼ ìí´ ëì¼í í´ë¬ì¤í°ë§ ê²°ê³¼ ëì¶ì© random_state=0
km_cluster = KMeans(n_clusters=5, max_iter=10000, random_state=0)
km_cluster.fit(feature_vect)
cluster_label = km_cluster.labels_
cluster_centers = km_cluster.cluster_centers_
```

> ê° êµ°ì§ë³ íµì¬ ë¨ì´ ì°¾ê¸°

- ndarrayì `argsort()[:,::â1]` :  cluster_centers ë°°ì´ ë´ ê°ì´ í° ìì¼ë¡ ì ë ¬ë ìì¹ ì¸ë±ì¤ ê°ì ë°í



```python
# êµ°ì§ë³ top n íµì¬ ë¨ì´, ê·¸ ë¨ì´ì ì¤ì¬ ìì¹ ìëê°, ëì íì¼ëª ë°í
def get_cluster_details(cluster_model, cluster_data, feature_names, clusters_num,
                        top_n_features=10):
  cluster_details = {}

  # cluster_centers arrayì ê°ì´ í° ìì¼ë¡ ì ë ¬ë ì¸ë±ì¤ ê°ì ë°í
  # êµ°ì§ ì¤ì¬ì (centroid)ë³ í ë¹ë word í¼ì²ë¤ì ê±°ë¦¬ê°ì´ í° ìì¼ë¡ ê°ì êµ¬íê¸° ìí¨
  centroid_feature_ordered_ind = cluster_model.cluster_centers_.argsort()[:,::-1]

  # ê°ë³ êµ°ì§ë³ë¡ ë°ë³µíë©´ì íµì¬ ë¨ì´, ê·¸ ë¨ì´ì ì¤ì¬ ìì¹ ìëê°, ëì íì¼ëª ìë ¥
  for cluster_num in range(clusters_num):
    # ê°ë³ êµ°ì§ë³ ì ë³´ë¥¼ ë´ì ë°ì´í° ì´ê¸°í.
    cluster_details[cluster_num] = {}
    cluster_details[cluster_num]['cluster'] = cluster_num

    # cluster_centers_.argsort()[:,::-1]ë¡ êµ¬í ì¸ë±ì¤ë¥¼ ì´ì©í´ top n í¼ì² ë¨ì´ë¥¼ êµ¬í¨
    top_feature_indexes = centroid_feature_ordered_ind[cluster_num, :top_n_features]
    top_features = [ feature_names[ind] for ind in top_feature_indexes ]

    # top_feature_indexesë¥¼ ì´ì©í´ í´ë¹ í¼ì² ë¨ì´ì ì¤ì¬ ìì¹ ìëê° êµ¬í¨
    top_feature_values = cluster_model.cluster_centers_[cluster_num,
                                                        top_feature_indexes].tolist()

    # cluster_details ëìëë¦¬ ê°ì²´ì ê°ë³ êµ°ì§ë³ íµì¬ë¨ì´ì ì¤ì¬ìì¹ ìëê°, í´ë¹ íì¼ëª ìë ¥
    cluster_details[cluster_num]['top_features'] = top_features
    cluster_details[cluster_num]['top_features_value'] = top_feature_values
    filenames = cluster_data[cluster_data['cluster_label'] == cluster_num]['filename']
    filenames = filenames.values.tolist()

    cluster_details[cluster_num]['filenames'] = filenames

  return cluster_details
```
